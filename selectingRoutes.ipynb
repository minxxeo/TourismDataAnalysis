```python
import folium
import pandas as pd
from geopy.distance import geodesic
from itertools import combinations

# 식당 간의 거리를 계산하여 2km 이내인 경우 그룹화
groups = []
group_id = 0
group_map = {}

for i, row1 in miryang_sample.iterrows():
    for j, row2 in miryang_sample.iterrows():
        if i < j:
            dist = geodesic((row1['식당위도'], row1['식당경도']), (row2['식당위도'], row2['식당경도'])).km
            if dist <= 2:
                if i in group_map:
                    group_map[j] = group_map[i]
                elif j in group_map:
                    group_map[i] = group_map[j]
                else:
                    group_map[i] = group_id
                    group_map[j] = group_id
                    group_id += 1

# 그룹화되지 않은 나머지 식당들도 각각의 그룹으로 추가
for i in range(len(miryang_sample)):
    if i not in group_map:
        group_map[i] = group_id
        group_id += 1

# Folium 지도에 그룹별로 표시
m = folium.Map(location=[miryang_sample['식당위도'].mean(), miryang_sample['식당경도'].mean()], zoom_start=12)

colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray']

for i, row in miryang_sample.iterrows():
    group = group_map[i]
    folium.Marker(
        [row['식당위도'], row['식당경도']], #changed from row['식당위'] to row['식당위도']
        popup=f"{row['식당명']} ({row['식당대표전화번호']})",
        icon=folium.Icon(color=colors[group % len(colors)])
    ).add_to(m)

# 지도 출력
m
```

```python
import requests
import folium
from IPython.display import display
import pandas as pd

# Directions API 호출 함수
def get_directions(df, origin_idx, destination_idx, waypoints_idx=[]):
    origin = df.iloc[origin_idx]
    destination = df.iloc[destination_idx]

    # 요청 데이터 설정
    request_params = {
        "origin": f"{origin['식당경도']},{origin['식당위도']}",
        "destination": f"{destination['식당경도']},{destination['식당위도']}",
        "waypoints": "|".join([f"{df.iloc[idx]['식당경도']},{df.iloc[idx]['식당위도']}" for idx in waypoints_idx]),
        "priority": "RECOMMEND",
        "car_fuel": "GASOLINE",
        "car_hipass": "false",
        "alternatives": "false",
        "road_details": "false"
    }

    headers = {
        "Authorization": f"KakaoAK {REST_API_KEY}"
    }

    response = requests.get(
        "https://apis-navi.kakaomobility.com/v1/directions",
        params=request_params,
        headers=headers
    )

    if response.status_code == 200:
        directions_data = response.json()
        return directions_data
    else:
        return response.status_code, response.text

# 지도 생성 및 경로 표시 함수
def create_map_with_route(df, origin_idx, destination_idx, waypoints_idx=[]):
    directions = get_directions(df, origin_idx, destination_idx, waypoints_idx)

    if isinstance(directions, tuple):
        print(f"Error: {directions[0]} - {directions[1]}")
        return None

    # Folium 지도 생성
    m = folium.Map(location=[df['식당위도'].mean(), df['식당경도'].mean()], zoom_start=12)

    # 출발지, 경유지, 도착지 마커 추가
    folium.Marker(
        [df.iloc[origin_idx]['식당위도'], df.iloc[origin_idx]['식당경도']],
        popup=f"출발지: {df.iloc[origin_idx]['식당명']}",
        icon=folium.Icon(color='green')
    ).add_to(m)

    for i, idx in enumerate(waypoints_idx):
        folium.Marker(
            [df.iloc[idx]['식당위도'], df.iloc[idx]['식당경도']],
            popup=f"경유지 {i+1}: {df.iloc[idx]['식당명']}",
            icon=folium.Icon(color='blue')
        ).add_to(m)

    folium.Marker(
        [df.iloc[destination_idx]['식당위도'], df.iloc[destination_idx]['식당경도']],
        popup=f"도착지: {df.iloc[destination_idx]['식당명']}",
        icon=folium.Icon(color='red')
    ).add_to(m)

    # 경로 폴리라인 추가
    if 'routes' in directions:
        for section in directions['routes'][0]['sections']:
            for road in section['roads']:
                if isinstance(road, dict) and 'vertexes' in road:
                    vertexes = road['vertexes']
                    if isinstance(vertexes, list) and vertexes:  # 비어있지 않는지!!
                        points = [(vertexes[i+1], vertexes[i]) for i in range(0, len(vertexes), 2)]
                        folium.PolyLine(points, color="blue", weight=2.5, opacity=1).add_to(m)

    return m

# 예시 호출
origin_idx = 0  # 출발지
destination_idx = 1  # 목적지
waypoints_idx = [2, 3]  # 경유지

# 지도 생성
map_with_route = create_map_with_route(miryang_sample, origin_idx, destination_idx, waypoints_idx)

# Colab에서 지도 표시
if map_with_route:
    display(map_with_route)
```
